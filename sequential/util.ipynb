```python
from pydub import AudioSegment
import os

input_folder = "mix"
output_folder = "mix"

# Create the output folder if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Scan the input folder for MP3 files
for filename in os.listdir(input_folder):
    if filename.endswith(".mp3"):
        mp3_path = os.path.join(input_folder, filename)
        wav_path = os.path.join(output_folder, os.path.splitext(filename)[0] + ".wav")

        # Open the MP3 file using pydub
        audio = AudioSegment.from_mp3(mp3_path)

        # Export the audio to WAV file
        audio.export(wav_path, format="wav")

        print(f"Converted {filename} to WAV")

print("Conversion complete.")
```

```python
import os
import shutil

source_directory = '../speech-sep-metrics/results/standard_model/audio_results'

destination_directory = 'audio_results'

# Iterate over the entries in the source directory
for entry in os.scandir(source_directory):
    if entry.is_file() and "_mix.wav" in entry.name:
        source_file = entry.path
        destination_file = os.path.join(destination_directory, entry.name)
        shutil.copy2(source_file, destination_file)
        print(f'File copied: {source_file} -> {destination_file}')
```
